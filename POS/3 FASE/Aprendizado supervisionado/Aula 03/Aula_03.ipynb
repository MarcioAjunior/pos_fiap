# Modelos de classsificação

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from sklearn.datasets import load_iris
X, y  = load_iris(return_X_y=True, as_frame=True)

df = X
df['target'] = y
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
df.head()
```

```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['species']
```

```python
# Separando em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3 ,random_state=42)
```

### Regressão logistica (classificador logistico)

```python
from sklearn.linear_model import LogisticRegression

#Treinando o modelo
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
#Relizando predições
y_pred = lr.predict(X_test)
```

```python
#Verificando métricas
print(f'Acurácia = {accuracy_score(y_test, y_pred)}')
print(f'precision_score = {precision_score(y_test, y_pred, average="micro")}')
print(f'recall_score = {recall_score(y_test, y_pred, average="micro")}')
print(f'f1_score = {f1_score(y_test, y_pred, average="micro")}')
```

### KNN K vizinhos mais proximos

```python
from sklearn.neighbors import KNeighborsClassifier

#Treinando o modelo
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
```

```python
#Relizando predições
y_pred = knn.predict(X_test)
```

```python
#Verificando métricas
print(f'Acurácia = {accuracy_score(y_test, y_pred)}')
print(f'precision_score = {precision_score(y_test, y_pred, average="micro")}')
print(f'recall_score = {recall_score(y_test, y_pred, average="micro")}')
print(f'f1_score = {f1_score(y_test, y_pred, average="micro")}')
```

```python
number_neighbords = range(1,106)
dt_accuracy_score = {}

for i in number_neighbords:
    knn_model = KNeighborsClassifier(n_neighbors = i)
    knn_model.fit(X_train, y_train)
    knn_pred = knn_model.predict(X_test)
    dt_accuracy_score[i] = accuracy_score(y_test, knn_pred)

dt_accuracy_score
```

```python
plt.plot(number_neighbords, dt_accuracy_score.values())
plt.xlabel('Number of Neightbords')
plt.ylabel('Accuracy')
plt.show()
```

### Arvores de descição

```python
from sklearn.tree import DecisionTreeClassifier

#Treinando o modelo
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
#Relizando as predições
y_pred = dtc.predict(X_test)
```

```python
#Verificando métricas
print(f'Acurácia = {accuracy_score(y_test, y_pred)}')
print(f'precision_score = {precision_score(y_test, y_pred, average="micro")}')
print(f'recall_score = {recall_score(y_test, y_pred, average="micro")}')
print(f'f1_score = {f1_score(y_test, y_pred, average="micro")}')
```

```python
from sklearn import tree

tree.plot_tree(dtc,
               feature_names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
               class_names=['setosa', 'virginica', 'versicolor'],
               impurity = False,
               label = 'none',
               filled=True
               )
```

### SVM support vector machines

```python
y_pred = svc.predict(X_test)
```
