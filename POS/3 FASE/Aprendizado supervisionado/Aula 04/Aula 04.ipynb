# Modelos de regressão

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
#metricas
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error, root_mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
#modelos
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
df = pd.read_csv('house_data.csv')
df.head()
```

```python
df.shape
```

```python
df = df[['price', 'bedrooms', 'bathrooms', 'sqft_lot', 'floors', 'waterfront']]
```

```python
df.head()
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
#Padronizando a escala dos valores numéricos do dataset
min_max_scaler = StandardScaler() 
X = min_max_scaler.fit_transform(X)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3 ,random_state=42)
```

### Regressão linear

```python
#Trinando o modelos
lr = LinearRegression()
lr.fit(X_train, y_train)

#Predizendo
y_pred = lr.predict(X_test)
```

```python
mse = root_mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test,y_pred)

print(f'mse = {mse}')
print(f'rmse = {rmse}')
print(f'mape = {mape}')
```

### KNN regressor

```python
knr = KNeighborsRegressor(n_neighbors=7, metric='euclidean')
```

```python
knr.fit(X_train, y_train)
y_pred = knr.predict(X_test)
```

```python
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f'mape = {mape}')
```

### SVM regressor SVR (support vector machine para regressão)

```python
from sklearn.svm import SVR

svr = SVR()
svr.fit(X_train, y_train)

y_pred = svr.predict(X_test)
```

```python
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f'mape = {mape}')
```

```python
#Passando dois hiperparâmetros
svr = SVR(kernel='linear', C=100)
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f'mape = {mape}')
```

# Arvore de regressão

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
```

```python
y_pred = dtr.predict(X_test)
```
